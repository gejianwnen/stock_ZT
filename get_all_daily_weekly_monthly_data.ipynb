```python
# get daily data
start_date = "20100101"
end_date = "20200430"
save_dir = "./data/stock/daily/"

num_err = 0
if GET_DAILY:
    for i in range(len(pool_df)):
        stock_code = pool_df.loc[i,"ts_code"]
        stock_name = pool_df.loc[i,"name"]
        if True:# stock_code[0] == "0":
            if i%100 == 0:
                print('正在获取第%d家: %s,股票代码%s.' % (i+1,stock_name, stock_code))
            #接口限制访问200次/分钟，加一点微小的延时防止被ban
            time.sleep(0.333)
            df = df = ts.pro_bar(ts_code = stock_code, asset='E', adj = "qfq", freq = "D", factors = ["tor","vr"],
                                 start_date = start_date, end_date = end_date )
            try:
                df.to_csv(save_dir + stock_code + "-"+ stock_name + '.csv', index = False)
            except Exception as err:
                print("err: ",stock_code,stock_name,err)
                try:
                    df.to_csv(save_dir + stock_code + '.csv', index = False)
                except Exception as err:
                    print("err: ",stock_code,stock_name,err)
                    num_err = num_err +1
```
