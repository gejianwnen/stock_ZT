# select by Index and stock

```python
def selectByKdj(df, index_monthly_MACD = True, index_weekly_KDJ = True, stock_monthly_KDJ = True, stock_weekly_KDJ = True ):
    select_df = df.copy(deep = True)
    select_df = select_df.loc[select_df["kdj_signal"] == 1, :]
    if index_monthly_MACD:
        select_df = select_df.loc[select_df["dif_monthly_index"].values >= select_df['dea_monthly_index'].values, :]
    else:
        select_df = select_df.loc[select_df["dif_monthly_index"].values < select_df['dea_monthly_index'].values, :]
    if index_weekly_KDJ:
        select_df = select_df.loc[select_df["K_weekly_index"].values >= select_df['D_weekly_index'].values, :]
    else:
        select_df = select_df.loc[select_df["K_weekly_index"].values < select_df['D_weekly_index'].values, :]
    if stock_monthly_KDJ:
        select_df = select_df.loc[select_df["K_monthly"].values >= select_df['D_monthly'].values, :]
    else:
        select_df = select_df.loc[select_df["K_monthly"].values < select_df['D_monthly'].values, :]
    if stock_weekly_KDJ:
        select_df = select_df.loc[select_df["K_weekly"].values >= select_df['D_weekly'].values, :]
    else:
        select_df = select_df.loc[select_df["K_weekly"].values < select_df['D_weekly'].values, :]
    
    return select_df
```

```python
stock_code = "000001.SZ"
start_date = "20100101"
end_date = "20200314"
save_dir = "./output/select_by_KDJ/"

index_monthly_MACD = True
index_weekly_KDJ = not True
stock_monthly_KDJ = True
stock_weekly_KDJ = True
if index_monthly_MACD:
    save_dir +=  "index_monthly_MACD_good/"
else:
    save_dir +=  "index_monthly_MACD_bad/"
if index_weekly_KDJ:
    save_dir +=  "index_weekly_KDJ_good/"
else:
    save_dir +=  "index_weekly_KDJ_bad/"
if stock_monthly_KDJ:
    save_dir +=  "stock_monthly_KDJ_good/"
else:
    save_dir +=  "stock_monthly_KDJ_bad/"
if stock_weekly_KDJ:
    save_dir +=  "stock_weekly_KDJ_good/"
else:
    save_dir +=  "stock_weekly_KDJ_bad/"
print(save_dir)

index_df = getIndex("000001.SH",start_date, end_date  , LOCAL = True)
pool_df = getStockBasic(LOCAL = True)
all_select_df = pd.DataFrame()
for i in range(30):
    stock_code = pool_df.loc[i,"ts_code"]
    stock_name = pool_df.loc[i,"name"]
    if stock_code[0] == "3":
        continue
    if stock_name[0] == "*":
        continue
        
    stock_df = getStock(stock_code,start_date, end_date  , LOCAL = True)
    all_df = mergeStockIndex(stock_df,index_df)
    if len(all_df) > 300:
        select_df = selectByKdj(all_df, index_monthly_MACD, index_weekly_KDJ, stock_monthly_KDJ , stock_weekly_KDJ )
        print(stock_name,stock_code,select_df.shape)
        if len(select_df) > 0:
            all_select_df  = pd.concat([all_select_df,select_df],axis = 0)
            for i in range(len(select_df)):
                ix = select_df.index[i]
                trade_date = select_df.trade_date.values[i]
                # print(ix)
                # print(trade_date)
                display_df = displaySelect(all_df,ix, trade_date= trade_date, save_dir = save_dir )
                plt.close()
all_select_df.to_csv(save_dir+"all_select_df.csv", index = False)
print(all_select_df.shape)
```

```python
def selectByStockKdj(df, stock_monthly_KDJ = True, stock_weekly_KDJ = True ):
    select_df = df.copy(deep = True)
    select_df = select_df.loc[select_df["kdj_signal"] == 1, :]
    select_df = select_df.loc[select_df["D_monthly"].values < 50, :]
    select_df = select_df.loc[select_df["D_weekly"].values < 60, :]
    select_df = select_df.loc[select_df["D"].values < 50, :]
    if stock_monthly_KDJ:
        select_df = select_df.loc[select_df["K_monthly"].values >= select_df['D_monthly'].values, :]
    else:
        select_df = select_df.loc[select_df["K_monthly"].values < select_df['D_monthly'].values, :]
    if stock_weekly_KDJ:
        select_df = select_df.loc[select_df["K_weekly"].values >= select_df['D_weekly'].values, :]
    else:
        select_df = select_df.loc[select_df["K_weekly"].values < select_df['D_weekly'].values, :]
    
    return select_df
```
