```python
GET_FINANCE = True
```

```python
start_date = '20010101'
end_date = '20200313'
trade_code  =  '600000.SH'

finance_df = pro.income(ts_code= trade_code, start_date=start_date, end_date=end_date)
# ann_date  	公告日期
# f_ann_date 	实际公告日期
# end_date  	报告期
finance_df.sort_values(by="end_date", ascending=True,inplace=True)
print(finance_df.shape)
finance_df.head()
```

```python
# get daily data
startdate = '20010101'
enddate = '20200313'
save_dir = "./data/stock/finance/income/"

pool_df = pro.stock_basic()

if GET_FINANCE:
    for i in range(len(pool_df)):
        stock_code = pool_df.loc[i,"ts_code"]
        stock_name = pool_df.loc[i,"name"]
        if i%100 == 0:
            print('正在获取第%d家: %s,代码%s.' % (i+1,stock_name, stock_code))
        #接口限制访问200次/分钟，加一点微小的延时防止被ban
        time.sleep(0.3)
        df =  pro.income(ts_code= trade_code, start_date=start_date, end_date=end_date)
        try:
            df.to_csv(save_dir + stock_code + "-"+ stock_name + '.csv', index = False)
        except:
            try:
                df.to_csv(save_dir + stock_code + '.csv', index = False)
            except:
                print("err: ", stock_name)
```

```python
# get daily data
startdate = '20010101'
enddate = '20200313'
save_dir = "./data/stock/finance/balance/"

pool_df = pro.stock_basic()

if GET_FINANCE:
    for i in range(len(pool_df)):
        stock_code = pool_df.loc[i,"ts_code"]
        stock_name = pool_df.loc[i,"name"]
        if i%100 == 0:
            print('正在获取第%d家: %s,代码%s.' % (i+1,stock_name, stock_code))
        #接口限制访问200次/分钟，加一点微小的延时防止被ban
        time.sleep(0.3)
        df =  pro.balancesheet(ts_code= trade_code, start_date=start_date, end_date=end_date)
        try:
            df.to_csv(save_dir + stock_code + "-"+ stock_name + '.csv', index = False)
        except:
            try:
                df.to_csv(save_dir + stock_code + '.csv', index = False)
            except:
                print("err: ", stock_name)
```
